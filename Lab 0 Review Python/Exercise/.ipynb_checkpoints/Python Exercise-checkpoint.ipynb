# Exercise 1

```python
total_score = 0
for i in range(1, 5):
    score = float(input(f"Enter the score for subject {i}: "))
    total_score = total_score + score
###########################################
print(f"Total score: {total_score}")
###########################################
# Determining the grade based on the average score
if total_score >= 90:
    grade = 'A'
elif 80 < total_score < 90:
    grade = 'B'
elif 60 <= total_score <= 80:
    grade = 'C'
else:
    grade = 'D'

# Displaying the grade
print("final grade is " + grade)
```

# Exercise 2

```python
char = input("Enter a character: ")
vowels = {'a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U'}
###########################################################
if char in vowels:
    print(f"The character '{char}' is a vowel.")
else:
    print(f"The character '{char}' is not a vowel.")
```

```python
#############################################################
# you can use lower() or upper() built in function #########
############################################################
char = input("Enter a character: ").lower()
vowels = {'a', 'e', 'i', 'o', 'u'}
###########################################################
if char in vowels:
    print(f"The character '{char}' is a vowel.")
else:
    print(f"The character '{char}' is not a vowel.")
```

# Exercise 3

```python
input_string = input("Enter a string: ").lower()
# Reversing the string
reversed_string = input_string[::-1]
######################################
reversed_string = reversed(input_string) 
# Checking if the string is a palindrome
if input_string == reversed_string:
    print(f"The string '{input_string}' is a palindrome.")
else:
    print(f"The string '{input_string}' is not a palindrome.")
```

# Exercise 4

```python
s1 = input("Enter the first string (s1): ")
s2 = input("Enter the second string (s2): ")

# Find the middle index of s1
middle_index = len(s1) // 2

# Split s1 into two halves
first_half = s1[:middle_index]
second_half = s1[middle_index:]

# Create the new string by appending s2 in the middle of s1
new_string = first_half + s2 + second_half

# Printing the new string
print(f"The new string is: {new_string}")
```

# Exersice 5

```python
input_string = input("Enter a string: ")

char_count = {}

for char in input_string:
    # If the character is already in the dictionary, increment its count
    if char in char_count:
        char_count[char] += 1
    # Otherwise, add the character to the dictionary with a count of 1
    else:
        char_count[char] = 1

# Printing the character counts
print(char_count)

for char, count in char_count.items():
    print(f"'{char}': {count}")
```

# Exercise 6 

```python
number = int(input("Enter a number: "))

# Initializing the factorial to 1
factorial = 1

# Checking if the number is negative, zero, or positive
if number < 0:
    print("Factorial does not exist for negative numbers.")
elif number == 0:
    print("The factorial of 0 is 1.")
########################################################
else:
    for i in range(1, number + 1):
        factorial *= i
    print(f"The factorial of {number} is {factorial}.")
```

# Exercise 7

```python
n = int(input("Enter a natural number: "))

sum_of_squares = 0

for i in range(1, n + 1):
    sum_of_squares += i * i

# Printing the result
print(f"The sum of the squares of the first {n} natural numbers is: {sum_of_squares}")
```

# Exercise 8 

```python
num = int(input("Enter a number: "))

sum = 0

# Calculate the sum of numbers from 1 to num
for i in range(1, num + 1):
    sum += i

# Print the result
print(f"The sum of all numbers from 1 to {num} is: {sum}")
```

# Exercise 9

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = []

# Iterate through the list and square each number
for num in numbers:
    squared_numbers.append(num ** 2)

# Print the squared numbers
print("Original numbers:", numbers)
print("Squared numbers:", squared_numbers)
```

```python
numbers = [1, 2, 3, 4, 5]

# Using list comprehension to square each number
squared_numbers = [x ** 2 for x in numbers]

# Print the squared numbers
print("Original numbers:", numbers)
print("Squared numbers:", squared_numbers)
```

# Exercise 10

```python
my_list = [10, 20, 30, 20, 40]

# Check if 20 is present and replace the first occurrence with 200
if 20 in my_list:
    index = my_list.index(20)
    my_list[index] = 200

# Print the updated list
print("Updated list:", my_list)
```

# Exercise 11

```python
my_list = [10, 20, 30, 20, 40]

# Iterate through the list and remove all occurrences of 20
while 20 in my_list:
    my_list.remove(20)

# Print the updated list
print("Updated list:", my_list)
```

```python
my_list = [10, 20, 30, 20, 40]

# Remove all occurrences of 20 using list comprehension
my_list = [x for x in my_list if x != 20]

# Print the updated list
print("Updated list:", my_list)
```

# Exercise 12

```python
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
tuple3 = (7, 8, 9)

result = []

for i in range(len(tuple1)):
    sum_value = tuple1[i] + tuple2[i] + tuple3[i]
    result.append(sum_value)

# Convert the result list to a tuple
result_tuple = tuple(result)

# Print the result
print("Element-wise sum of tuples:", result_tuple)
```

# Exercise 13

```python
keys = ['a', 'b', 'c']
values = [1, 2, 3]

# Initialize an empty dictionary
my_dict = {}

# Iterate over the indices of keys (assuming keys and values have the same length)
for i in range(len(keys)):
    my_dict[keys[i]] = values[i]

# Print the dictionary
print("Dictionary:", my_dict)
```

```python
keys = ['a', 'b', 'c']
values = [1, 2, 3]

# Using dict() constructor with zip
my_dict = dict(zip(keys, values))

# Print the dictionary
print("Dictionary:", my_dict)
```
