```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import f_classif, SelectKBest
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.manifold import TSNE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df=pd.read_csv("test.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df=df.drop("method",axis=1)
```

```python
# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=5) # Set perplexity less than number of samples
df_new = tsne.fit_transform(df)
```

```python
# Save the results in new file

import numpy as np
# Assuming 'df_new' contains the t-SNE transformed data
np.savetxt("tsne_results.csv", df_new, delimiter=",")
```

```python
#  ؛lot the df_new , need index number in every point from 1 to 22

import matplotlib.pyplot as plt

# Scatter plot with index numbers
plt.figure(figsize=(8, 6))
for i in range(df_new.shape[0]):
    plt.scatter(df_new[i, 0], df_new[i, 1])
    plt.text(df_new[i, 0], df_new[i, 1], str(i+1))  # Add index number as text

plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE Visualization with Index Numbers')
plt.show()
```
