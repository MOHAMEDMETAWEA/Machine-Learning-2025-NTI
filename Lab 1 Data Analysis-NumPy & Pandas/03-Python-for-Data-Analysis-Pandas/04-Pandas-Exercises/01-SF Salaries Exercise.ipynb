** What is the name of highest paid person (including TotalPayBenefits)?**

```python
# Find the employee with the highest TotalPay
highest_pay_row = data.loc[data['TotalPay'].idxmax()]
print(highest_pay_row)
```

** What is the name of lowest paid person (including benefits)? Do you notice something strange about how much he or she is paid?**

```python
# ** What is the name of lowest paid person (including benefits)? Do you notice something strange about how much he or she is paid?**
lowest_pay_row = data.loc[data['TotalPay'].idxmin()]
print(lowest_pay_row)
```

** What was the average (mean) BasePay of all employees per year? (2011-2014) ? **

** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
# ** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **
job_counts_2013 = data[data['Year'] == 2013]['JobTitle'].value_counts()
unique_jobs_2013 = job_counts_2013[job_counts_2013 == 1].count()
print(unique_jobs_2013)
```

** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
#** How many people have the word Chief in their job title? (This is pretty tricky) **
def chief_string(title):
    if 'chief' in title.lower().split():
        return True
    else:
        return False
chief_count = data['JobTitle'].apply(chief_string).sum()
print(chief_count)
```

** Bonus: Is there a correlation between length of the Job Title string and Salary? **

```python
# ** Bonus: Is there a correlation between length of the Job Title string and Salary? **
data['title_len'] = data['JobTitle'].apply(len)
data[['title_len', 'TotalPayBenefits']].corr()
```
