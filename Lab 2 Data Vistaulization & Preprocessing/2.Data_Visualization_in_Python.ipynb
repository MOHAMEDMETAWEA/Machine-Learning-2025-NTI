
# Data Visualization in Python

Data visualization is a crucial step in data analysis and exploration. It helps to uncover patterns, trends, and insights by representing data in graphical form.

In this notebook, we will explore various types of visualizations using Python libraries: **Matplotlib** and **Seaborn**. We will use the `tips` dataset, which contains information about restaurant tips.

## Topics Covered:
1. Bar Plot
2. Count Plot
3. Distribution Plot
4. Heatmap
5. Scatter Plot
6. Box Plot
7. Pair Plot


```python

# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
data = sns.load_dataset('tips')

data.head()
```

```python
# Display the first few rows
data['day'].value_counts()
```

```python
data[['tip','day']]
```

```python
sunday_tips = data[data['day'] == 'Sun']['tip']
print("Tips on Sunday:")
print(sunday_tips)
```

```python
sunday_tips.describe()
```


## 1. Bar Plot

A **bar plot** is used to visualize the average value of a numerical variable for different categories.

### Example: Average Tip Amount by Day


```python

# Bar Plot: Average Tip Amount by Day
sns.barplot(x='day', y='tip', data=data, palette='muted')
plt.title('Average Tip Amount by Day')
plt.show()
```

The black line in the bar plot you uploaded represents an error bar, specifically showing the standard error or confidence interval for the tip amount on that day. The error bar gives an indication of the variability or uncertainty of the average tip amount.


## 2. Count Plot

A **count plot** displays the count of occurrences of each category in a categorical variable.

### Example: Count of Tips by Gender


```python

# Count Plot: Count of Tips by Gender
sns.countplot(x='sex', data=data, palette='pastel')
plt.title('Count of Tips by Gender')
plt.show()
```


## 3. Distribution Plot

A **distribution plot** shows the distribution of a numerical variable.

### Example: Distribution of Total Bill Amounts


```python

# Distribution Plot: Total Bill Amounts
sns.histplot(data['total_bill'], kde=True, color='blue', bins=20)
plt.title('Distribution of Total Bill Amounts')
plt.show()
```

kde=True: This adds a Kernel Density Estimate (KDE) curve to the histogram, showing the smoothed distribution of the data.

color='blue': Specifies that the color of the histogram bars will be blue.

bins=20: This divides the data into 20 bins (or "bars") for the histogram.


## 4. Heatmap

A **heatmap** displays data in a matrix format, where the values are represented by varying colors.

### Example: Correlation Heatmap for Numerical Variables


```python
data.head()
```

```python
# Drop rows with non numeric values in data

data = data[['total_bill','tip','size']]
```

```python
# Heatmap: Correlation between numerical variables
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```


## 5. Scatter Plot

A **scatter plot** is used to visualize the relationship between two numerical variables.

### Example: Relationship between Total Bill and Tip


```python
import seaborn as sns
data = sns.load_dataset('tips')
data.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='day', data=data, palette='viridis', s=100)
plt.title('Relationship between Total Bill and Tip')
plt.show()
```


## 6. Box Plot

A **box plot** shows the distribution of a numerical variable and highlights outliers.

### Example: Distribution of Total Bills by Day


```python

# Box Plot: Total Bill by Day
sns.boxplot(x='day', y='total_bill', data=data, palette='Set2')
plt.title('Distribution of Total Bills by Day')
plt.show()
```


## 7. Pair Plot

A **pair plot** shows pairwise relationships between numerical variables and can help visualize clusters or trends.

### Example: Pairwise Relationships in the Dataset


```python

# Pair Plot: Pairwise relationships
sns.pairplot(data, hue='sex', palette='husl')
plt.show()
```
