# Exercises

## 1. What is the distribution of passengers’ ages?

```python
# plot here
sns.histplot(titanic['age'].dropna(), bins=30, kde=True, color='blue')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## 2. What is the distribution of passengers’ ages and fare ?
each figure in subplot

```python
# plot here
plt.figure(figsize=(10,6))
sns.histplot(titanic['fare'].dropna(), bins=40, kde=True, color='green')
plt.title('Fare Distribution of Titanic Passengers')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

## 3. What is the relationship between age and fare?

```python
# plot here
#  relationship between age and fare?
sns.scatterplot(x='age', y='fare', data=titanic, hue='class', palette='Set2', s=100)
plt.title('Relationship between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

## 4. What is the relationship between survived and age?

```python
# plot here
#  relationship between survived and age

sns.scatterplot(x='survived', y='age', data=titanic, hue='survived', palette='Set1', s=100)
plt.title('Survived vs Age colored by Survival')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

## 5. How does the survival rate vary by gender?

```python
# get mean of survived when column of sex == male
mean = titanic[titanic['sex'] == 'male']['survived'].mean()
print("Mean survival rate for males:", mean)
```

```python
# get mean of survived when column of sex == female
mean = titanic[titanic['sex'] == 'female']['survived'].mean()
print("Mean survival rate for males:", mean)
```

```python
# plot here
sns.barplot(x='sex', y='survived', data=titanic, palette='pastel')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```
